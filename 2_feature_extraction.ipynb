```python
import pandas as pd
import re
from textblob import TextBlob
import nltk
from sklearn.feature_extraction.text import CountVectorizer
#from readability import Readability


# Download NLTK data for tokenization
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')

```

```python
df = pd.read_csv('data/News_dataset/initial_features.csv')
```

```python
df.head()
```

### Sentiment Polarity Score

```python
def sentiment_polarity(text):
    return TextBlob(text).sentiment.polarity

df['sentiment_polarity'] = df['text'].apply(sentiment_polarity)
```

### Subjectivity Score

```python
def subjectivity_score(text):
    return TextBlob(text).sentiment.subjectivity

df['subjectivity_score'] = df['text'].apply(subjectivity_score)
```

```python
df.head()
```

### Average Sentence Length

```python
def avg_sentence_length(text):
    sentences = nltk.sent_tokenize(text)
    if len(sentences) == 0:
        return 0
    return sum(len(nltk.word_tokenize(sentence)) for sentence in sentences) / len(sentences)

df['avg_sentence_length'] = df['text'].apply(avg_sentence_length)
```

### Character Count

```python
df['character_count'] = df['text'].apply(len)
```

### Exclamation Mark Count

```python
df['exclamation_mark_count'] = df['text'].apply(lambda x: x.count('!'))
```

### Question Mark Count

```python
df['question_mark_count'] = df['text'].apply(lambda x: x.count('?'))
```

### Capitalized Word Count

```python
def capitalized_word_count(text):
    return sum(1 for word in nltk.word_tokenize(text) if word.isupper())

df['capitalized_word_count'] = df['text'].apply(capitalized_word_count)

df.head()
```

```python
df.to_csv('data/News_dataset/final_features.csv', index = False)
```
